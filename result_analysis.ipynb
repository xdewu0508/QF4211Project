## Table 4: Transaction Cost Drag

```python
df = pd.read_csv("result/paper_tables/cost_drag_summary_by_asset_model_strategy_feasible.csv")

# aggregate across assets
a6 = (
    df.groupby(["Model", "Strategy Mode"], as_index=False)
      .agg({
          "Sharpe Cost Drag": "mean",
          "Avg Return Cost Drag": "mean",
          "Cumulative Return Cost Drag": "mean"
      })
)
a6
```
