## Appendix B

### Table 1: Sample of Feasible Strategy Configurations (With Cost)

```python
df = pd.read_csv("result/paper_tables/table_detail_feasible_configs.csv")
df = df[df["Cost Regime"] == "with_cost"]
print(df.columns)
df = df.drop(columns=["Model Family", "Input Type", "Feature Type", "Threshold Feasible", "Feasibility Bucket", "Model Params"])

# sort by performance
df = df.sort_values(by="Annualized Sharpe", ascending=False)
# show top 10
df.head(10)
```

### Table 2: Comparison of Strategy Performance Before and After Transaction Costs

```python
df = pd.read_csv("result/paper_tables/cost_comparison_by_configuration_feasible.csv")

# keep only relevant columns
b2 = df[[
    "Asset",
    "Model",
    "Strategy Mode",
    "Test Sharpe no_cost",
    "Test Sharpe with_cost",
    "Cumulative Return no_cost",
    "Cumulative Return with_cost"
]].copy()

# compute cost impact explicitly (nice addition)
b2["Sharpe Cost Drag"] = b2["Test Sharpe no_cost"] - b2["Test Sharpe with_cost"]
b2["Return Cost Drag"] = b2["Cumulative Return no_cost"] - b2["Cumulative Return with_cost"]

# sort by largest Sharpe deterioration
b2 = b2.sort_values(by="Sharpe Cost Drag", ascending=False)
# show top 10 rows
b2 = b2.head(10)
b2
```

### Table 3: Input Representation Comparison (OHLC vs Candlestick)

```python
df = pd.read_csv("result/paper_tables/table10_feasible_bundle.csv")

# apply filters
b3 = df[
    (df["Scope"] == "all_assets") &
    (df["Cost_Regime"] == "with_cost")
].copy()

# select relevant columns
b3 = b3[[
    "Input Type",
    "Strategy_Mode",
    "Test Sharpe",
    "Test Precision",
    "Test Avg Return"
]]

# rename for cleaner presentation
b3 = b3.rename(columns={
    "Strategy_Mode": "Strategy Mode",
    "Test Sharpe": "Annualized Sharpe",
    "Test Avg Return": "Average Return"
})
# sort for readability
b3 = b3.sort_values(by=["Strategy Mode", "Annualized Sharpe"], ascending=[True, False])

b3
```

### Table 4: Feature Engineering Comparison (Raw vs Extended Features)

```python
df = pd.read_csv("result/paper_tables/table11_feasible_bundle.csv")

# apply filters
b4 = df[
    (df["Scope"] == "all_assets") &
    (df["Cost_Regime"] == "with_cost")
].copy()

# select relevant columns
b4 = b4[[
    "Feature Type",
    "Strategy_Mode",
    "Test Sharpe",
    "Test Precision",
    "Test Avg Return"
]]

# rename for clean presentation
b4 = b4.rename(columns={
    "Strategy_Mode": "Strategy Mode",
    "Test Sharpe": "Annualized Sharpe",
    "Test Avg Return": "Average Return"
})
b4 = b4.sort_values(by=["Strategy Mode", "Annualized Sharpe"], ascending=[True, False])

b4
```
