```python
df = pd.read_csv("result/summary_all_strategies_feasible.csv")

# check columns
print(df.columns)

# sort by Sharpe for each asset
df = df[(df["Cost Regime"] != "no_cost") & (df["Model Family"] == "ML") & (df["Strategy Label"] == "long")]
df = df.sort_values(["Asset", "Annualized Sharpe"], ascending=[True, False])
df = df[["Asset", "Model", "Strategy Label", "Annualized Sharpe"]]
df[df["Asset"] == "BTC"]["Annualized Sharpe"].mean()
df.head()
```
